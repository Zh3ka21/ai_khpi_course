```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
# Завантажуємо набір даних
iris_df = datasets.load_iris()
# Описуємо модель
model = KMeans(n_clusters=3)

# Проводимо моделювання
model.fit(iris_df.data)

# Отримуємо центри кластерів
cluster_centers = model.cluster_centers_
# Отримуємо мітки кластерів для кожного зразка в наборі даних
labels = model.labels_
print(labels)
# Візуалізуємо дані
plt.figure(figsize=(12, 6))
# Візуалізація за двома ознаками (приклад: перший і другий стовпці у наборі даних)
plt.subplot(1, 2, 1)
plt.scatter(iris_df.data[:, 0], iris_df.data[:, 1], c=labels)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker='o', s=200, c='red', label='Центри кластерів')

#Can be changed 0 -> 3
plt.xlabel(iris_df.feature_names[0])
plt.ylabel(iris_df.feature_names[1])
plt.legend()

# Візуалізація за іншими двома ознаками (приклад: третій і четвертий стовпці у наборі даних)
plt.subplot(1, 2, 2)
plt.scatter(iris_df.data[:, 2], iris_df.data[:, 3], c=labels)
plt.scatter(cluster_centers[:, 2], cluster_centers[:, 3], marker='o', s=200, c='red', label='Центри кластерів')
plt.xlabel(iris_df.feature_names[2])
plt.ylabel(iris_df.feature_names[3])
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.utils import shuffle
from scipy.cluster.hierarchy import linkage, dendrogram

iris = load_iris()
data, target = shuffle(iris.data, iris.target)

samples = data[:50]
species = target[:50]

mergings = linkage(samples, method='complete')

# Строим дендрограмму с именами видов справа
dendrogram(mergings,
           labels=iris.target_names[species],
           leaf_rotation=50,
           leaf_font_size=6,
           )

plt.show()
```

```python
from sklearn import datasets
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt


iris_df = datasets.load_iris()

model = TSNE(learning_rate=1, n_components=3) 

transformed = model.fit_transform(iris_df.data)

x_axis = transformed[:, 0]
y_axis = transformed[:, 1]
z_axis = transformed[:, 2]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(x_axis, y_axis, z_axis, c=iris_df.target, cmap=plt.cm.get_cmap('viridis', 3))

plt.show()
```

```python
# Імпортуємо бібліотеки
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA

# Завантажуємо датасет
iris = load_iris()

# Визначаємо модель
dbscan = DBSCAN(eps=1, min_samples=15)

# Навчаємо
dbscan.fit(iris.data)

# Зменшуємо розмірність за допомогою метода PCA - головних компонент
pca = PCA(n_components=2).fit(iris.data)
pca_2d = pca.transform(iris.data)

# Будуємо у відповідності до трьох класів
for i in range(0, pca_2d.shape[0]):
    if dbscan.labels_[i] == 0:
        c1 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='r', marker='+')
    elif dbscan.labels_[i] == 1:
        c2 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='g', marker='o')
    elif dbscan.labels_[i] == -1:
        c3 = plt.scatter(pca_2d[i, 0], pca_2d[i, 1], c='b', marker='*')

plt.legend([c1, c2, c3], ['Кластер 1', 'Кластер 2', 'Шум'])
plt.title('DBSCAN знайшов 2 кластери та шум')
plt.show()
```
